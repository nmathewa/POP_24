```python
# add the background map

import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.feature as cfeature

```

```python
# Create a figure object 

eraland = xr.open_dataset('/media/nmathewa/nma_backup/Datasets/Datasets/era5_land/sensible_2014_dec_era5L/data.nc')
fig = plt.figure(figsize=(10,10))

# create an axis 

ax = fig.subplots(subplot_kw={'projection': ccrs.PlateCarree()})

# plot the data
eraland['sshf'].isel(time=0).plot(ax=ax, transform=ccrs.PlateCarree())

ax.coastlines()
#ax.set_global()
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, facecolor='lightgrey')
ax.add_feature(cfeature.OCEAN)
```
